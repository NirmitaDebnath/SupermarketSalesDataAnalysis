**EXPLORATORY DATA ANALYSIS**

**Q.WHAT IS THE OVERALL SALES TREND?**

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Date'].min()
```

```python
df['Date'].max()
```

```python
df['month-year']=df['Date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
df['month']=df['Date'].apply(lambda x: x.strftime('%m'))
```

```python
df['month']
```

```python
df['month-year']
```

```python
df.groupby(['month'],as_index=False)['SELLING PRICE'].sum()
```

```python
sns.barplot(x='month',y='SELLING PRICE',data=df.groupby(['month'],as_index=False)['SELLING PRICE'].sum(),palette='Paired')
plt.show()
```

```python
df.groupby(['month'],as_index=False)['Quantity'].sum()
```

```python
sns.barplot(x='month',y='Quantity',data=df.groupby(['month'],as_index=False)['Quantity'].sum(),palette='Paired')
plt.show()
```
